```python
# Nettoyage initial de csv Apec

apec = apec.drop(columns = "Unnamed: 0")
apec = apec.rename(columns={'Expérience': 'Experience', 'Profil recherché' : 'Profil recherche'})

def remove_numbers_hyphens_from_end(s):
    if len(s) <= 7:
        return re.sub(r'[\d-]', '', s)
    else:
        # Split into two parts: everything except the last 7 characters, and the last 7 characters
        main_part = s[:-7]
        end_part = s[-7:]
        # Remove numbers and hyphens from the last 7 characters
        end_part = re.sub(r'[\d-]', '', end_part)
        # Combine the parts back together
        s = main_part + end_part
        s = s.strip()
        return s

# Apply the function to 'column1'
apec['Localisation'] = apec['Localisation'].apply(remove_numbers_hyphens_from_end)

def remove_numbers_from_start(s):
    if len(s) <= 3:
        return re.sub(r'^\d+', '', s)
    else:
        # Split into two parts: the first 3 characters and the rest of the string
        start_part = s[:3]
        main_part = s[3:]
        # Remove numbers from the first 3 characters
        start_part = re.sub(r'\d', '', start_part)
        # Combine the parts back together
        s = start_part + main_part
        s = s.strip()
        return s

# Apply the function to 'column1'
apec['Contrat'] = apec['Contrat'].apply(remove_numbers_from_start)

apec['Date de parution'] = apec['Date de parution'].str.replace('Publiée le ', '', regex=False)

apec.replace('A négocier', np.nan, inplace=True)

apec.head()
```

```python
# Nettoyage et transformation de la colonne 'Date de debut'

def parse_and_transform_date(date_str):
    if pd.notna(date_str) and ' ' in date_str:
        try:
            date_str = date_str.replace('janvier', '01')
            date_str = date_str.replace('février', '02')
            date_str = date_str.replace('mars', '03')
            date_str = date_str.replace('avril', '04')
            date_str = date_str.replace('mai', '05')
            date_str = date_str.replace('juin', '06')
            date_str = date_str.replace('juillet', '07')
            date_str = date_str.replace('août', '08')
            date_str = date_str.replace('septembre', '09')
            date_str = date_str.replace('octobre', '10')
            date_str = date_str.replace('novembre', '11')
            date_str = date_str.replace('décembre', '12')
            
            parsed_date = datetime.strptime(date_str, '%d %m %Y').strftime('%d/%m/%Y')
            return parsed_date
        except ValueError:
            return date_str
    else:
        return date_str

df['Date de debut'] = df['Date de debut'].apply(parse_and_transform_date)
```

```python
# Nettoyage et transformation de la colonne 'Date de parution'

def parse_date(date_str):
    try:
        parsed_date = parse(date_str, dayfirst=True)
        return parsed_date.strftime('%d/%m/%Y')
    except:
        return pd.NaT

df['Date de parution'] = df['Date de parution'].apply(parse_date)
```
