```python
# Nettoyage initial de csv Apec

apec = pd.concat([apec, apec], ignore_index=True)

apec = apec.drop(columns = "Unnamed: 0")
apec = apec.rename(columns={'Expérience': 'Experience', 'Profil recherché' : 'Profil recherche'})

def remove_numbers_hyphens_from_end(s):
    if len(s) <= 7:
        return re.sub(r'[\d-]', '', s)
    else:
        # Split into two parts: everything except the last 7 characters, and the last 7 characters
        main_part = s[:-7]
        end_part = s[-7:]
        # Remove numbers and hyphens from the last 7 characters
        end_part = re.sub(r'[\d-]', '', end_part)
        # Combine the parts back together
        s = main_part + end_part
        s = s.strip()
        return s

# Apply the function to 'column1'
apec['Localisation'] = apec['Localisation'].apply(remove_numbers_hyphens_from_end)

def remove_numbers_from_start(s):
    if len(s) <= 3:
        return re.sub(r'^\d+', '', s)
    else:
        # Split into two parts: the first 3 characters and the rest of the string
        start_part = s[:3]
        main_part = s[3:]
        # Remove numbers from the first 3 characters
        start_part = re.sub(r'\d', '', start_part)
        # Combine the parts back together
        s = start_part + main_part
        s = s.strip()
        return s

# Apply the function to 'column1'
apec['Contrat'] = apec['Contrat'].apply(remove_numbers_from_start)

apec['Date de parution'] = apec['Date de parution'].str.replace('Publiée le ', '', regex=False)

apec.replace('A négocier', np.nan, inplace=True)

apec.head()
```

```python
# Nettoyage et transformation de la colonne 'Date de debut'

# Change 'Non renseigné' par missing value
df['Date de debut'] = df['Date de debut'].replace('Non renseigné', np.nan)

def parse_and_transform_date(date_str):
    if pd.notna(date_str) and ' ' in date_str:
        try:
            date_str = date_str.replace('janvier', '01')
            date_str = date_str.replace('février', '02')
            date_str = date_str.replace('mars', '03')
            date_str = date_str.replace('avril', '04')
            date_str = date_str.replace('mai', '05')
            date_str = date_str.replace('juin', '06')
            date_str = date_str.replace('juillet', '07')
            date_str = date_str.replace('août', '08')
            date_str = date_str.replace('septembre', '09')
            date_str = date_str.replace('octobre', '10')
            date_str = date_str.replace('novembre', '11')
            date_str = date_str.replace('décembre', '12')
            
            parsed_date = datetime.strptime(date_str, '%d %m %Y').strftime('%d/%m/%Y')
            return parsed_date
        except ValueError:
            return date_str
    else:
        return date_str

df['Date de debut'] = df['Date de debut'].apply(parse_and_transform_date)
```

```python
# Nettoyage et transformation de la colonne 'Date de parution'

df['Date de parution'] = df['Date de parution'].replace('Non renseigné', np.nan)

df['Date de parution'] = df['Date de parution'].apply(parse_and_transform_date)

def parse_date(date_str):
    try:
        parsed_date = parse(date_str, dayfirst=True)
        return parsed_date.strftime('%d/%m/%Y')
    except:
        return pd.NaT

df['Date de parution'] = df['Date de parution'].apply(parse_date)
```

```python
# Division de titres en métiers plus prrécis

keywords = {
    'Data Analyst': ['analyst', 'analysis', 'analytics', 'bi', 'analyse', 'données', 'donée', 'dataviz', 'analytique'],
    'Data Scientist': ['scientist', 'science', 'machine learning', 'ml', 'ai', 'artificial intelligence', 'intelligence artificielle', 'ia'],
    'Data Business': ['business', 'strategist', 'strategy', 'consultant', 'analytics manager'],
    'Data Engineer': ['engineer', 'engineering', 'etl', 'data pipeline', 'data warehousing', 'architect', 'ingénieur', 'ingenieur', 'intégration', 'intégrateur', 'deep learning'],
    'Manager Data': ['manager', 'director', 'head of', 'chief', 'leader', 'management', 'gestion', 'master', 'responsable'],
    'Chef de projet data': ['chef de projet', 'chef', 'projet', 'owner', 'scrum'],
    'Développeur Data': ['développeur'],
    'Data quality': ['quality', 'qualité', 'auditeur', 'audit'],
    'Staticticien': ['statisticien', 'statistiques', 'statistique', 'statisticien(ne)'],
}

# Fonction pour attribuer une catégorie basée sur le job_title
def categorize_job(title):
    title = title.lower()
    for category, words in keywords.items():
        if any(word in title for word in words):
            return category
    return 'Other'

# Créer une nouvelle colonne job_cluster
df['Job category'] = df['Poste'].apply(categorize_job)

# Vérifier et supprimer les 'other' métiers s'ils correspondent pas
df = df.loc[~df['Job category'].str.contains('Other')]
```

```python
# Définir les mappings pour regrouper les termes similaires

skill_mappings = {
    "Analyse": ["Analysis", "Analytics", "Analyse"],
    "Machine learning": ["Apprentissage Automatique", "Machine learning", "ML"],
    "Intelligence Artificielle": ["Intelligence Artificielle", "IA", "AI", "Artificial Intelligence"]
}

# Créer une table de correspondance inverse pour la recherche rapide
reverse_skill_mappings = {term: main for main, variations in skill_mappings.items() for term in variations}

# Définir les soft skills et hard skills en utilisant des listes prédéfinies
soft_skills = [
    "Communication", "Innovation", "Collaboration", "Autonomie", "Curiosité", "Efficacité", "Gestion de Projet",
    "Initiative", "Leadership", "Motivation", "Flexibilité", "Créativité", "Adaptabilité", "Prise de Décision",
    "Négociation", "Gestion des Risques", "Résolution de Problèmes", "Priorisation", "Résilience", "Communication Orale",
    "Travail d'Équipe", "Ouverture d'Esprit", "Communication Écrite", "Compétences Analytiques", "Diplomatie", "Indépendance",
    "Empathie", "Persuasion", "Gestion du Temps", "Attention aux Détails", "Écoute Active", "Gestion des Tâches",
    "Gestion d'Équipe", "Coordination de Projet", "Apprentissage Continu", "Compétences Organisationnelles", 
    "Prise de Décision Stratégique", "Désir d'Apprendre", "Leadership d'Équipe", "Pensée Analytique", 
    "Collaboration en Équipe", "Construction de Relations", "Gestion de la Pression", "Compétences en Présentation",
    "Pensée Critique", "Interprétation de Données"
]

hard_skills = [
    "Analysis", "Analytics", "Dataviz", "Analyse", "SQL", "Excel", "Tableau", "Power BI", "R", "Python", "Analyse de Données",
    "Apprentissage Automatique", "Gestion de Projet", "Nettoyage de Données", "Visualisation de Données", "Google Analytics", "scikit-learn",
    "TensorFlow", "Keras", "Deep learning", "Traitement du Langage Naturel", "NLP", "Data mining", "Big Data", "Hadoop", "Spark",
    "Outils BI", "CRM", "Modélisation des Processus Métier", "Entrepôt de Données", "Data warehouse", 
    "Datalake", "ETL", "Flux de Données", "Data pipelines", "Plateformes Cloud", "AWS", "Azure", "GCP", "Architecture de Données",
    "Gouvernance des Données", "Gestion des Risques", "KPI", "Méthodologies Agiles", "Intégration", 
    "Data integration", "Audit", "Assurance Qualité", "QA", "Validation des Données", "Traçabilité des Données",
    "Analyse de Régression", "Analyse des Séries Chronologiques", "Analyse Statistique", "Processus ETL", 
    "API", "Machine learning", "ML", "AI", "Artificial Intelligence", "Intelligence Artificielle", "IA"
]

# Extraire les skills
def extract_skills(text, skills_list):
    if pd.isna(text):
        return set()
    found_skills = set()
    text_lower = text.lower()
    for skill in skills_list:
        if re.search(r'\b' + re.escape(skill.lower()) + r'\b', text_lower, re.IGNORECASE):
            found_skills.add(skill)
    # Normaliser les compétences trouvées en utilisant le mapping inverse
    normalized_skills = {reverse_skill_mappings.get(skill, skill) for skill in found_skills}
    return normalized_skills

def combine_skills(description, profile, competences, skills_list):
    combined_skills = (extract_skills(description, skills_list) |
                       extract_skills(profile, skills_list) |
                       extract_skills(competences, skills_list))
    return ", ".join(sorted(combined_skills))

# Créer de nouvelles colonnes pour les hard skills et soft skills en extrayant depuis 'Description', 'Profil recherche' et 'Competences'
df['Hard Skills'] = df.apply(lambda row: combine_skills(row['Description'], row['Profil recherche'], row['Competences'], hard_skills), axis=1)
df['Soft Skills'] = df.apply(lambda row: combine_skills(row['Description'], row['Profil recherche'], row['Competences'], soft_skills), axis=1)

df['Hard Skills'] = df['Hard Skills'].replace('', 'Non renseigné')
df['Soft Skills'] = df['Soft Skills'].replace('', 'Non renseigné')
```

```python
# Définir la faonction pour nrttoyer le texte
def clean_text_french(text):
    if not isinstance(text, str):  # Vérifie si la valeur est une chaîne de caractères
        return ""  # Retourne une chaîne vide si ce n'est pas le cas
    
    # Tokeniser le texte
    doc = nlp(text.lower())
    
    # Retirer la ponctuation et les mots vides (stopwords)
    tokens = [token.text for token in doc if not token.is_stop and not token.is_punct and token.is_alpha]
    
    # Lemmatize (optionnel, si vous préférez utiliser spaCy pour lemmatisation)
    # tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and token.is_alpha]
    
    return ' '.join(tokens)

# Appliquer la fonction de nettoyage
df['Cleaned_Description'] = df['Description'].apply(clean_text_french)
```
